## Trabalho de Estatística II

Análise econômica sobre diferença de genêros

### Objetivo
---
O objetivo desta rotina é analisar se o número de desocupação e de empregadores em percentuais possui
uma diferença signficativa entre gêneros e correlacionar isto ao número médio de horas semanais dedicadas aos cuidados de pessoas e/ou afazeres domésticos.

> Importação das bibliotecas Python

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Dados sobre empregadores
---
Observa-se diferença signficativa no número histórico de mínimos e máximos percentuais entre empregadores
separados por genêro.

O máximo percentual dentre os anos analisados é bem maior para homens enquanto o efeito
inverso (mínimo percentual mais relevante) é relacioanado às mulheres, o que indica uma relação de
diferença entre homens e mulheres que nos mostra que estas estão em menos posições de liderança e 
protagonismo empresarial no quesito empreendedorismo, apesar de ambos seguirem uma linha de tendência parecida.

```python
file_path = './Tabela-1.6.xlsx'
excel_data = pd.ExcelFile(file_path)

# Cria uma lista para armazenar os dados de cada aba
data = []

# Itera sobre cada aba do arquivo
for sheet_name in excel_data.sheet_names:
    # Verifica se "CV" não está no nome da aba
    if "CV" not in sheet_name:
        # Lê apenas a aba atual
        sheet_df = excel_data.parse(sheet_name)
        
        # Extrai os valores das células C10 e D10
        homens = sheet_df.iloc[8, 2]  # Linha 10, coluna C
        mulheres = sheet_df.iloc[8, 3]  # Linha 10, coluna D
        
        # Adiciona os dados à lista
        data.append({'Ano': sheet_name, 'Homens': homens, 'Mulheres': mulheres})

# Converte a lista em um DataFrame
dfA = pd.DataFrame(data)

# Exibe o DataFrame
display(dfA)

## Análise de mínimos e máximos por sexo
minimos = dfA[['Homens', 'Mulheres']].min()
maximos = dfA[['Homens', 'Mulheres']].max()
print('MÍNIMOS PERCENTUAIS (11 anos): \n', minimos)
print('MÁXIMOS PERCENTUAIS (11 anos): \n', maximos)


dfA['Ano'] = pd.to_numeric(dfA['Ano'], errors='coerce')

# Ordena o DataFrame pelo ano
dfA = dfA.sort_values(by='Ano')

# Cria o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(dfA['Ano'], dfA['Homens'], label='Homens', marker='o')
plt.plot(dfA['Ano'], dfA['Mulheres'], label='Mulheres', marker='o')
plt.title('Proporção de Empregadores por Ano (Homens e Mulheres)')
plt.xlabel('Ano')
plt.ylabel('Proporção de empregadores (%)')
plt.legend()
plt.grid(True)
plt.show()
```

### Participação na força de trabalho
---
Observa-se diferença signficativa no número histórico de mínimos e máximos percentuais entre empregadores
separados por genêro.

```python
file_path = './Tabela-1.3.xlsx'
excel_data = pd.ExcelFile(file_path)

# Cria uma lista para armazenar os dados de cada aba
data = []

# Itera sobre cada aba do arquivo
for sheet_name in excel_data.sheet_names:
    # Verifica se "CV" não está no nome da aba
    if "CV" not in sheet_name:
        # Lê apenas a aba atual
        sheet_df = excel_data.parse(sheet_name)
        
        # Extrai os valores das células C10 e D10
        homens = sheet_df.iloc[6, 2]  # Linha 10, coluna C
        mulheres = sheet_df.iloc[6, 3]  # Linha 10, coluna D
        
        # Adiciona os dados à lista
        data.append({'Ano': sheet_name, 'Homens': homens, 'Mulheres': mulheres})

# Converte a lista em um DataFrame
dfB = pd.DataFrame(data)

# Exibe o DataFrame
display(dfB)
print(dfB.describe())
## Análise de mínimos e máximos por sexo
minimos = dfB[['Homens', 'Mulheres']].min()
maximos = dfB[['Homens', 'Mulheres']].max()
print('MÍNIMOS PERCENTUAIS (11 anos): \n', minimos)
print('MÁXIMOS PERCENTUAIS (11 anos): \n', maximos)

# Converte a coluna 'Ano' para numérica, se necessário
dfB['Ano'] = pd.to_numeric(dfB['Ano'], errors='coerce')

# Ordena o DataFrame pelo ano
dfB = dfB.sort_values(by='Ano')

# Cria o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(dfB['Ano'], dfB['Homens'], label='Homens', marker='o')
plt.plot(dfB['Ano'], dfB['Mulheres'], label='Mulheres', marker='o')
plt.title('Proporção de Pessoas Ocupadas por Ano (Homens e Mulheres)')
plt.xlabel('Ano')
plt.ylabel('Proporção Ocupada (%)')
plt.legend()
plt.grid(True)
plt.show()
```
